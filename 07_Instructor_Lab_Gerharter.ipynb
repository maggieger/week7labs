#### List of train-test splits for 8-fold cross-validation

```python
from sklearn.model_selection import KFold
```

```python
kf = KFold(n_splits=8)  

for fold, (train_idx, test_idx) in enumerate(kf.split(gh_teams), start=1):
    train_data = gh_teams.iloc[train_idx]
    test_data = gh_teams.iloc[test_idx]
    
    print(f"Fold {fold}")
    print(f"Train data shape: {train_data.shape}")
    print(f"Test data shape: {test_data.shape}")
    print("-" * 40)
```
